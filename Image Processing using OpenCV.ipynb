```python
import cv2 # Importing the OpenCV library for image processing

from google.colab import files # Importing files module for uploading file
from google.colab.patches import cv2_imshow # Importing the cv2_imshow for displaying images in Colab

print("Upload an image")
image=files.upload() # Uploading the image
for fn in image.keys(): # Gets the filename of the uploaded image
  filename = fn
  break

read=cv2.imread(filename) # Reads the image file using OpenCV
cv2_imshow(read) # Displays the original image in Colab
print()

height = 300
width = int(read.shape[1] * height / read.shape[0]) # Calculates the width based on the desired height and aspect ratio

resized_image = cv2.resize(read, (width, height)) # Resizes the image to the calculated dimensions
cv2_imshow(resized_image) # Displays the resized image in Colab
print()

rotated_image = cv2.rotate(resized_image, cv2.ROTATE_180) # Rotates the resized image by 180 degrees
cv2_imshow(rotated_image) # Displays the rotated image in Colab
print()
```
